```python
import shutil, os
# ZIP_PATH = "/content/drive/MyDrive/hama/hama1.zip"
# OUTPUT_BASE = "/content/drive/MyDrive/dataset/pests"   # permanen di GDrive
# TEMP_EXTRACT = "/content/temp_extract"                 # folder sementara
ZIP_PATH = r"C:\Users\Admin\Downloads\hama1.zip"
OUTPUT_BASE = r"D:\6S2\Thesis\7 dataset\hama1"   
TEMP_EXTRACT = r"D:\6S2\Thesis\7 dataset\hama1\temp_extract"                
```

```python


# Hapus folder output jika sudah ada
if os.path.exists(OUTPUT_BASE):
    shutil.rmtree(OUTPUT_BASE)

# Hapus folder temp extract jika ada
if os.path.exists(TEMP_EXTRACT):
    shutil.rmtree(TEMP_EXTRACT)

# Buat ulang folder
os.makedirs(OUTPUT_BASE, exist_ok=True)
os.makedirs(TEMP_EXTRACT, exist_ok=True)
```

```python
import zipfile

with zipfile.ZipFile(ZIP_PATH, 'r') as z:
    z.extractall(TEMP_EXTRACT)

print("Selesai extract ZIP ke folder sementara.")
```

```python
import os
import shutil

PREP_FOLDER = os.path.join(TEMP_EXTRACT, "preprocessedDataset")

class_counts = {}

# Loop setiap folder class (0–11)
for class_name in sorted(os.listdir(PREP_FOLDER)):
    src_dir = os.path.join(PREP_FOLDER, class_name)
    dst_dir = os.path.join(OUTPUT_BASE, class_name)

    os.makedirs(dst_dir, exist_ok=True)

    files = sorted(os.listdir(src_dir))
    class_counts[class_name] = len(files)

    # Rename file 000.jpg, 001.jpg, dst
    for i, fname in enumerate(files):
        ext = fname.split(".")[-1].lower()
        new_name = f"{i:03d}.{ext}"  # format 000, 001, 002

        shutil.copy(os.path.join(src_dir, fname),
                    os.path.join(dst_dir, new_name))

print("Proses pemindahan & penamaan selesai!")
```

```python
import pandas as pd

total_images = sum(class_counts.values())

df = pd.DataFrame({
    "Class": list(class_counts.keys()),
    "Jumlah Gambar": list(class_counts.values())
})

print("=== Informasi Dataset ===")
print(df)
print("\nTotal gambar:", total_images)
```
